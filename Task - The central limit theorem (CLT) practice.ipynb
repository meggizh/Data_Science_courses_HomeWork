# The central limit theorem (CLT) practice on the example of exponential continuous distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import expon
%matplotlib inline
```

An example of the dependence of the probability density of an exponential distribution on a parameter Scale (Lambda)

```python
x = np.arange(0, 5, 0.0001)
plt.plot(x, expon.pdf(x,scale=1/1.5),'r-')
plt.plot(x, expon.pdf(x,scale=1),'g-')
plt.plot(x, expon.pdf(x,scale=1/0.5),'b-')
plt.ylabel('P(x)')
plt.xlabel('$x$')
```

Sample volume 1000:

```python
sc=1/0.5
exp_rv = expon.rvs(scale=sc,size=1000)
```

Sample histogram and theoretical distribution density of a random variable:

```python
plt.plot(x, expon.pdf(x, scale=sc),
         'r-', lw=5, alpha=0.8, label='expon pdf=distribution density')
plt.hist(exp_rv, [0, 1, 2, 3, 4, 5], density=True, label='hist=Sample histogram')
plt.ylabel('P(x), frequency')
plt.xlabel('$x$')
plt.legend(loc='best')
```

Theoretical mean and variance of a random variable:

```python
exp_median = expon.median(scale=sc)
exp_mean = expon.mean(scale=sc)
exp_var = expon.var(scale=sc)
exp_std = expon.std(scale=sc)
exp_expect = expon.expect(scale=sc)
print('exp_median =',exp_median, 
      '; exp_mean =',exp_mean, 
      '; Disp = exp_var =', exp_var, 
      '; exp_std =', exp_std,
      '; expect =', exp_expect)
```

For n = 5, 10, 50, we generate 1000 samples of size n and build histograms of the distributions of their sample means, as well as the distribution density:

```python
# define a function that generates 1000 samples of n1 sample means

def task(n1):
    sample_n1= np.arange(start=0, stop=1000, step=1,dtype=float)
    for i in range(0,1000):
        exp_rv_n1 = expon.rvs(scale=sc,size=n1)
        sample_n1[i]=sum(exp_rv_n1)/(float(n1))
    return sample_n1
```

```python
# n=5
plt.hist(task(5), bins=15, density=True, label='Sample histogram n=5')
norm_rv1 = sts.norm(exp_expect, (exp_var/float(5))**(0.5))
pdf1 = norm_rv1.pdf(x)
plt.plot(x, pdf1, label='distribution density')
plt.ylabel('frequency')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
# n=10
plt.hist(task(10), bins=15, density=True, label='Sample histogram n=10')
norm_rv1 = sts.norm(exp_expect, (exp_var/float(10))**(0.5))
pdf1 = norm_rv1.pdf(x)
plt.plot(x, pdf1, label='distribution density')
plt.ylabel('frequency')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
# n=50
plt.hist(task(50), bins=15, density=True, label='Sample histogram n=50')
norm_rv1 = sts.norm(exp_expect, (exp_var/float(50))**(0.5))
pdf1 = norm_rv1.pdf(x)
plt.plot(x, pdf1, label='distribution density')
plt.ylabel('frequency')
plt.xlabel('$x$')
plt.legend(loc='best')
```

### Conclusion: the accuracy of approximating the distribution of sample means  increases with increasing n
